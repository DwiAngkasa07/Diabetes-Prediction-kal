```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/diabetes_prediction_dataset.csv')
data1
```

```python
# @title gender

from matplotlib import pyplot as plt
import seaborn as sns
data1.groupby('gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title age

from matplotlib import pyplot as plt
data1['age'].plot(kind='hist', bins=20, title='age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title hypertension

from matplotlib import pyplot as plt
data1['hypertension'].plot(kind='hist', bins=20, title='hypertension')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title heart_disease

from matplotlib import pyplot as plt
data1['heart_disease'].plot(kind='hist', bins=20, title='heart_disease')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title bmi

from matplotlib import pyplot as plt
data1['bmi'].plot(kind='hist', bins=20, title='bmi')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title smoking_history

from matplotlib import pyplot as plt
import seaborn as sns
data1.groupby('smoking_history').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title age vs hypertension

from matplotlib import pyplot as plt
data1.plot(kind='scatter', x='age', y='hypertension', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title smoking_history vs hypertension

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data1['smoking_history'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data1, x='hypertension', y='smoking_history', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title gender vs hypertension

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data1['gender'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data1, x='hypertension', y='gender', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title smoking_history vs age

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data1['smoking_history'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data1, x='age', y='smoking_history', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title gender vs smoking_history

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['smoking_history'].value_counts()
    for x_label, grp in data1.groupby('gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('gender')
_ = plt.ylabel('smoking_history')
```
